```python
import logging
import os
import sqlite3
import pandas as pd
from sqlalchemy import create_engine, inspect, text
```

```python
engine = create_engine("sqlite:///Data/sdm-dwh/sdmdb.db")

inspector = inspect(engine)
tables = inspector.get_table_names()

data = {}

# alles verwijderen
with engine.begin() as conn:
    for table in tables:
        try:
            conn.execute(text(f"DELETE FROM {table}"))
        except Exception as e:
            print(f"fout bij delete {table}: {e}")

# check
for table in tables:
    try:
        query = f"SELECT * FROM {table}"
        data[table] = pd.read_sql_query(query, engine)

        print(f"{table}: {len(data[table])} records")

    except Exception as e:
        print(f"fout bij load {table}: {e}")

print("\n✅ klaar: reset + reload compleet")
```
