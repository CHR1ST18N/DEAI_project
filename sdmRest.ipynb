```python
import pandas as pd
from sqlalchemy import create_engine, inspect

# =========================
# 1. CONNECTIE MAKEN
# =========================

engine = create_engine("sqlite:///sdmdb.db")
inspector = inspect(engine)

# =========================
# 2. ALLE TABELLEN OPHALEN
# =========================

tables = inspector.get_table_names()

# =========================
# 3. ALLE TABELLEN INLADEN
# =========================

data = {}

for table in tables:
    try:
        data[table] = pd.read_sql_table(table, engine)
        
    except Exception as e:
        print(f"❌ fout bij {table}: {e}")

# =========================
# 4. VOORBEELD GEBRUIK
# =========================

# bv. één tabel gebruiken:
# df_customer = data["customer"]

print("✅ klaar: alle tabellen ingeladen")
```
